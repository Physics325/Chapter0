```python
import matplotlib.pyplot as plt
from numpy import sin, exp, linspace
%matplotlib inline
```

```python
x = linspace(0.0,8.0,300)
```

```python
plt.plot(x,x**4*exp(-2*x),label="(a)")
plt.plot(x,(x**2*exp(-x)*sin(x**2))**2,label="(b)")
plt.legend()
plt.title("Problem 0-0")
plt.xlabel("x")
plt.ylabel("y")
```

```python
from numpy import log, log10
```

```python
counts,time = loadtxt('../files/Ba137.txt', unpack=True)
```

```python
plt.plot(time,counts)
```

```python
plt.semilogy(time,counts)
plt.title("Log scale")
plt.ylabel("Counts")
plt.xlabel("Time (s)")
```

### Or:

```python
plt.plot(time,log(counts))
plt.title("Log of raw data")
plt.ylabel("log(counts)")
plt.xlabel("Time (s)")
```

```python
N=30000
l=1/240
plt.semilogy(time,counts,"r.")
plt.semilogy(time,N*exp(-l*time),"b")
plt.title("Log of raw data")
plt.ylabel("log(counts)")
plt.xlabel("Time (s)")
```
